```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

from catboost import CatBoostRegressor
from skopt import BayesSearchCV
from skopt.space import Integer, Real

import os
import json
import joblib
import matplotlib.pyplot as plt
```

```python
# Data
df = pd.read_csv("StudentsPerformance_modified.csv")

# On prédit math score, on exclut les autres scores
X = df.drop(columns=["reading score", "writing score", "math score"])
y_math = df["math score"]

# CatBoost categorical handling
cat_cols = X.select_dtypes(include="object").columns.tolist()
cat_idx = [X.columns.get_loc(c) for c in cat_cols]

X_train_m, X_test_m, y_train_m, y_test_m = train_test_split(
    X, y_math, test_size=0.2, random_state=42
)

cv = KFold(n_splits=5, shuffle=True, random_state=42)

base_math = CatBoostRegressor(
    loss_function="RMSE",
    random_seed=42,
    verbose=0
)

search_space = {
    "depth": Integer(3, 10),
    "learning_rate": Real(1e-3, 0.3, prior="log-uniform"),
    "l2_leaf_reg": Real(1.0, 20.0),
    "iterations": Integer(300, 2000),
    "subsample": Real(0.5, 1.0),
    "random_strength": Real(0.0, 2.0),
}

opt_math = BayesSearchCV(
    estimator=base_math,
    search_spaces=search_space,
    n_iter=35,
    cv=cv,
    scoring="r2",
    n_jobs=-1,
    random_state=42,
    verbose=0
)

opt_math.fit(X_train_m, y_train_m, cat_features=cat_idx)

best_model_math = opt_math.best_estimator_
best_params_math = opt_math.best_params_
best_cv_r2_math = float(opt_math.best_score_)

print(" Bayes opt OK (math score). Best CV R2:", best_cv_r2_math)
print("Best params:", best_params_math)

best_model_math.fit(X_train_m, y_train_m, cat_features=cat_idx)

pred_math = best_model_math.predict(X_test_m)

rmse_val = float(np.sqrt(mean_squared_error(y_test_m, pred_math)))

math_metrics = {
    "cv_best_r2": best_cv_r2_math,
    "r2_test": float(r2_score(y_test_m, pred_math)),
    "mae_test": float(mean_absolute_error(y_test_m, pred_math)),
    "rmse_test": rmse_val,
    "y_test_mean": float(np.mean(y_test_m)),
}

print("\n math score test metrics:", math_metrics)
```

```python
plt.figure()
plt.scatter(y_test_m, pred_math, alpha=0.6)
plt.xlabel("True score")
plt.ylabel("Predicted score")
plt.title("Math Score — True vs Pred")
mn = min(y_test_m.min(), pred_math.min())
mx = max(y_test_m.max(), pred_math.max())
plt.plot([mn, mx], [mn, mx])
plt.show()

plt.figure()
plt.scatter(pred_math, y_test_m - pred_math, alpha=0.6)
plt.axhline(0)
plt.xlabel("Predicted")
plt.ylabel("Residuals")
plt.title("Math Score — Residuals vs Pred")
plt.show()

plt.figure()
plt.bar(X.columns, best_model_math.feature_importances_)
plt.xticks(rotation=90)
plt.title("Feature Importance — math score")
plt.show()

# ---------- Export model ----------
joblib.dump(best_model_math, "artifacts/catboost_math_score.joblib")
print("Model exported to: artifacts/catboost_math_score.joblib")
```
